```python
import pandas as pd
user_usage = pd.read_csv("user_usage.csv")
user_device = pd.read_csv("user_device.csv")
devices = pd.read_csv("android_devices.csv")
devices.rename(columns={"Retail Branding": "manufacturer"}, inplace=True)
```

There are Three main methods, *Join, Merge, and Concat*. However, Concat is mainly for series and opperations in a single dimension, so we are not goinh to focus on it today. Lets first look at *Join*

## Join
Join is the simplest and the most intuitive method for combining two dataframe. It works by combining the data frames by row index. Let's take a look!

```python
print(user_usage.shape)
user_usage.head()
```

```python
print(user_device.shape)
user_device.head()
```

```python
result=user_usage.join(user_device,how='left',lsuffix='**')
print(result.shape)
result.head()
```

```python
result=user_usage.join(user_device,how='right',lsuffix='**')
print(result.shape)
result.head()
```

```python
result[240:245]
```

```python
result=user_usage.join(user_device,how='inner',lsuffix='**')
print(result.shape)
result.head()
#inner is the intersection of the two dataframes, 
#so the result is the same as the left join
```

```python
result=user_usage.join(user_device,how='outer',lsuffix='**')
print(result.shape)
result[239:244]
#outer is the union of the two dataframes, 
#so the result is the same as the right join
```

```python
result=user_usage.join(user_device,how='cross',lsuffix='**')
print(result.shape)
result.head()
#this parirs every row of the left dataframe with every row of the right dataframe,
#thus forming an n*m dataframe. 
#240*272=65280
```

## Merge
Merge is a more versatial, powerful form of join. It allows us to combine data sets with regards to columns

```python
result = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id')
print(result.shape)
result.head()
```

What is happending here? The join method defaults to how=left when it is applied to a data set, whereas merge defaults to inner, which is the intersection between the two dataframes. Lets take a look by counting the number of matching keys below. 

```python
a=user_device['use_id']
b=user_usage['use_id']
count=0
#counts the number of matching user ids
#as these are customer ids, we can assume no repeating elements in a given dataframe
for x in a:
    temp=False
    for y in b:
        if y==x:
            temp=True
            break
    if temp:
        count+=1
print(count)
```

```python
print(user_device.shape)
print(user_usage.shape)
print(user_device['use_id'].unique().shape)
user_usage['use_id'].unique().shape
```

But why is this different from join's how=inner? Because remember that Join uses indices and merge uses columns. left was the same as inner for join because 272>240, thus there were 240 instances of overlaps in row indicies.

Let's take a look at how all the other methods play out in merge 

```python
result = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id', how='left')
print(result.shape)
print(result[:2])
result[200:205]
```

```python
result = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id', how='right')
print(result.shape)
result[239:244]
```

```python
result.head()
```

```python
result = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id', how='outer', indicator=True)
result.shape
#(240+272)-159=353
```

Merge can also use indices like join specificying the options left_index=True, right_index=True
